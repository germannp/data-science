# Onyx Advent Calendar 2023

I was mainly interested in the different grinders for which they provide settings for each coffee. Beyond that I looked for some patterns in the recipes, without finding much.

The base recipe was always the same, 25gr coffee and 400ml water, poured as follows:
* Bloom with 50ml
* 0:30 strong center pour to 200ml
* 1:30 spiral pour to 300ml
* 2:00 spiral pour to 400ml

The other parameters we scrape from the website below.

```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import requests
import pandas as pd
import seaborn as sns
```

```python
%load_ext blackcellmagic
```

```python
page = requests.get("https://onyxcoffeelab.com/products/2023-advent-calendar")
soup = BeautifulSoup(page.text)
```

```python
recipes = []
for coffee in soup.find_all("div", {"class": "pop-left"}):
    data = {
        "Name": coffee.h2.string,
    }

    for span in coffee.find("div", {"class": "features"}).find_all("span"):
        feature, value = span.string.split(": ")
        if feature != "Origin":
            data[feature] = value

    setup_ps = coffee.find("div", {"class": "box-setup"}).find_all("p")
    data["Filter"] = setup_ps[0].string[11:]
    temperature_in_F = float(setup_ps[2].string.split()[-1][:-2])
    data["Temperature"] = (temperature_in_F - 32) / 1.8

    for line in coffee.find("div", {"class": "box-grind"}).p.text.split("\n"):
        grind, grinder = line.split(" ", maxsplit=1)
        if grinder == "um":
            grinder = "Microns"
        elif grinder == "EK43":
            grinder = "Mahlkonig EK43"
        elif grinder == "Baratza Vario/Forte":
            grinder = "Baratza Vario+"
        elif grinder == "2/3 Fellow Ode Gen 2":
            grinder = "Fellow Ode Gen 2"
        data[grinder] = grind
    data["Drain"] = (
        coffee.find("div", {"class": "box-brew"})
        .p.text.split("\n")[-1]
        .split()[0]
        .replace("~", "")
    )
    recipes.append(data)

recipes = pd.DataFrame(recipes)
recipes.index += 1
recipes["Elevation"] = recipes["Elevation"].str[:-5]
int_grinders = [
    "Microns",
    "Baratza Encore",
    "Comandante C40",
]
recipes[int_grinders] = recipes[int_grinders].astype(int)
float_grinders = [
    "Mahlkonig EK43",
    "Fellow Ode Gen 2",
]
recipes[float_grinders] = recipes[float_grinders].astype(float)
recipes["Drain"] = pd.to_timedelta("00:" + recipes["Drain"])
recipes.to_csv("data/onyx_coffee_advent_calendar_2023.csv")
recipes
```

```python
grinds = int_grinders + float_grinders

sns.pairplot(recipes[grinds])
plt.show()
```

```python
simple_recipes = recipes.copy()
simple_recipes["Process"] = simple_recipes["Process"].apply(
    lambda process: "Washed" if process == "Washed" else "Processed"
)
simple_recipes["Filter"] = simple_recipes["Filter"].apply(
    lambda filter_: "Cone filter" if "one" in filter_ else "Kalita filter"
)
simple_recipes["Drain [min]"] = simple_recipes["Drain"] / pd.to_timedelta("1min")
```

```python
def plot_correlation(x, y):
    for (process, filter_), group in simple_recipes.groupby(["Process", "Filter"]):
        plt.scatter(
            group[x],
            group[y],
            color="tab:blue" if process == "Washed" else "tab:orange",
            marker="v" if filter_ == "Cone filter" else "*",
            label=f"{process}, {filter_}",
        )
    plt.gca().set_xlabel(x)
    plt.gca().set_ylabel(y)
    plt.legend(frameon=False, bbox_to_anchor=(1.05, 1), loc="upper left")
    sns.despine()
    plt.show()


plot_correlation("Drain [min]", "Microns")
```

```python
plot_correlation("Drain [min]", "Temperature")
```

```python
def plot_strips(by):
    parameters = ["Temperature", "Microns", "Drain [min]"]
    fig, axs = plt.subplots(nrows=len(parameters))
    for ax, parameter in zip(axs, parameters):
        data = (
            simple_recipes[[by, parameter]]
            .melt(id_vars=by, var_name="Parameter")
            .rename(columns={"value": parameter})
        )
        sns.swarmplot(data=data, x=parameter, y=by, ax=ax)
        ax.set_ylabel("")

    sns.despine(fig)
    plt.tight_layout()
    plt.show()


plot_strips("Process")
```
